```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")

train = train.drop(['Name', 'Ticket','Cabin'], axis=1)
test = test.drop(['Name', 'Ticket','Cabin'], axis=1)

train.head()
```

```python
all_data = pd.concat((train.loc[:,'Pclass':'Embarked'],test.loc[:,'Pclass':'Embarked']))
all_data.head()
```

```python
all_data['Embarked']=all_data['Embarked'].fillna('S')
```

```python
sum(all_data['Embarked'].isnull())
```

```python
all_data.head()
```

```python
all_data['Sex']=all_data['Sex'].apply(lambda x: 1 if x=='female' else 0)
```

```python
all_data.head()
```

```python
all_data.info()
```

```python
all_data.describe()
```

```python
all_data['Age']=all_data['Age'].fillna(all_data['Age'].mean())
```

```python
all_data.info()
```

```python
all_data['Fare']=all_data['Fare'].fillna(all_data['Fare'].mean())
```

```python
all_data['FamilySize'] = all_data.Parch + all_data.SibSp + 1
```

```python
all_data=all_data.drop(['Parch', 'SibSp'], axis=1)
```

```python
all_data['Fare']=all_data['Fare']/all_data['Fare'].mean()
```

```python
all_data['Age']=all_data['Age']/all_data['Age'].mean()
```

```python
all_data['Embarked'] = all_data['Embarked'].map({'S': 0, 'C': 1, 'Q': 2}).astype(int)
```

```python
X_train = all_data[:train.shape[0]]
X_test = all_data[train.shape[0]:]
y = train.Survived
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, y)
accuracy = round(logreg.score(X_train, y) * 100, 2)
print(accuracy)
logreg_preds = logreg.predict(X_test)
```
